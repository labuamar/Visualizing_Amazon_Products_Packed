```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# Loading csv and viewing first 10 rows
dimensions_scraped= pd.read_csv('C:/Users/Ludicrous/Desktop/Data_Portfolio/amazon_dimensions_weight_102621_2.csv')
dimensions_scraped.head(10)
```

```python
# Checking for null values
dimensions_scraped.info()
```

```python
#Dropping rows with null values
dimensions_scraped = dimensions_scraped.dropna()
```

```python
#Checking to see that rows with null values have been dropped
dimensions_scraped.head(50)
```

```python
dimensions_scraped['V']=dimensions_scraped['Weight'].apply(lambda x: 0 if 'inches' in x else 1)
```

```python
dimensions_scraped.head(50)
```

```python
dimensions_scraped[['Dimensions','Weight']] = dimensions_scraped[['Dimensions','Weight']].where(dimensions_scraped['V'] == 1, dimensions_scraped[['Weight','Dimensions']].values)
```

```python
dimensions_scraped.head(50)
```

```python
patterndel = "(B0.*)"
filter = dimensions_scraped['Dimensions'].str.contains(patterndel)
dimensions_scraped = dimensions_scraped[~filter]
```

```python
dimensions_scraped.head(50)
```

```python
patterndel = "(B0.*)"
filter = dimensions_scraped['Weight'].str.contains(patterndel)
dimensions_scraped = dimensions_scraped[~filter]
```

```python
dimensions_scraped.head(50)
```

```python
patterndel = "A-1"
filter = dimensions_scraped['Weight'].str.contains(patterndel)
dimensions_scraped = dimensions_scraped[~filter]
```

```python
dimensions_scraped.head(50)
```

```python
#V=0 for pounds V=1 for ounces
dimensions_scraped['V']=dimensions_scraped['Weight'].apply(lambda x: 0 if 'pounds' in x else 1)
```

```python
dimensions_scraped['Weight'] = dimensions_scraped['Weight'].str.split(' ').str[0].astype(float)
```

```python
dimensions_scraped.head(10)
```

```python
# Converting the weight column from ounces and pounds to grams
dimensions_scraped['Weight'] = 453.5 * dimensions_scraped['Weight'].where(dimensions_scraped['V'] == 0, 0.062 * dimensions_scraped['Weight'])
```

```python
dimensions_scraped=dimensions_scraped.drop(columns="V")
```

```python
dimensions_scraped["Dimensions"]=dimensions_scraped["Dimensions"].str.strip("inches")
```

```python
dimensions_scraped[['length','height','width']]=dimensions_scraped["Dimensions"].str.split("x", expand=True).astype(float)
```

```python
dimensions_scraped=dimensions_scraped.drop(columns="Dimensions")
```

```python
dimensions_scraped["id"]=dimensions_scraped.index + 1
```

```python
dimensions_scraped.head()
```

```python
dimensions_scraped=dimensions_scraped.rename({'Weight':'weight'},axis=1)
```

```python
dimensions_scraped.to_csv('check.csv')
```
